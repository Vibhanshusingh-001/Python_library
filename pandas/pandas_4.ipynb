```python
data1=pd.DataFrame({
    "key1":[1,2,3,4,5],
    "key2":[75,8,9,10,11],
    "key3":[12,13,14,15,16]    
})
```

# perform merge operation

```python
pd.merge(data1,data2) # ye kewal intersection data ko show karega final result me 
# by defualt merge function me inner ko hi pakdata h
```

```python
pd.merge(data1,data2,how="left") # left merge 
# ye data1 ko left treat karega aur data2 ko right treat karega 
# jo bhi data pahle metion kiya h aapne wo left ban jayega baki right 
# left ka matlab sabkuch uthao left se aur phir map kardo data2 se map 
# aur agr nhi hai to NaN likh do 
```

```python
pd.merge(data1,data2,how="right",on= "key1") # isaka matlab key1 se mao karana h 
```

```python
pd.merge(data1,data2,how="outer")
```

```python
pd.merge(data1,data2,how="cross") # cross join matlab sabka combination bana do 
```
